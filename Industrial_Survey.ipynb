```python
fp = r'/content/india-polygon.shp'
map_df = gpd.read_file(fp) 
map_df_copy = gpd.read_file(fp)
map_df
```

```python
map_df.plot();
```

```python
st_nm = list(map_df['st_nm'])
response = [0,0,0,0,0,0,0,0,0,13,0,0,0,26,0,0,0,39,0,0,0,0,18,0,0,0,0,14,13,0,0,0,14,0,0,0,0]

df = pd.DataFrame(list(zip(st_nm,response)))
df.columns = ['st_nm', 'response']
df
```

```python
#Merging the data
merged = map_df.set_index('st_nm').join(df.set_index('st_nm'))
#merged['Count'] = merged['Count'].replace(np.nan, 0)
merged.head()
```

```python
#Create figure and axes for Matplotlib and set the title
fig, ax = plt.subplots(1, figsize=(10, 10))
ax.axis('off')

ax.set_title('Distribution of responces across tear-one cities of india', fontdict={'fontsize': '16', 'fontweight' : '10'})

# Plot the figure
merged.plot(column='response',cmap='Greys', linewidth=0.8, ax=ax, edgecolor='0',legend=True,markersize=[39.739192, -104.990337]);
```

```python
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
data_response = data_response.rename(columns = {'Company Name':'company_name'})
```

```python
# Create a string of all the company names
company_names = ' '.join(data_response['company_name'].astype(str))
```

```python
# Generate a word cloud
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = set(STOPWORDS), 
                min_font_size = 10).generate(company_names)

# Plot the word cloud
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```

```python
data_response = data_response.rename(columns = {'How frequently do you recruit into the company/organization?':'recrutment_pattern'})
```

```python
data_response_grouped = data_response.groupby(["sector", "recrutment_pattern"]).size().reset_index(name='counts')
sns.barplot(x="sector", y="counts", hue="recrutment_pattern", data=data_response_grouped)
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns


# Create a color palette
palette = dict(zip(data_response_grouped["recrutment_pattern"].unique(),
                   sns.color_palette("husl", len(data_response_grouped["recrutment_pattern"].unique()))))

# Create the chart
plt.figure(figsize=(10, 6))
plt.rcParams.update({'font.size': 14})
sns.barplot(x="sector", y="counts", hue="recrutment_pattern", data=data_response_grouped, palette=palette)
#plt.xticks(rotation=70)
plt.legend(bbox_to_anchor=(0, 1, 1, 0))
plt.show()
```
